### Applying the Machine learning algorithm (LInear regression) on AirQuality dataset available on UCI http://archive.ics.uci.edu/ml/datasets/Air+Quality

```python
# Importing basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the dataset
dataset=pd.read_excel(r'C:\Users\Hemant Raturi\Downloads\DA & ML\ML AirQualityUCI.xlsx')
dataset.head(50)
```

Since we have three values(Temprature, Relative humidity, Absolute humidity) to predict we will take three different prediction vector y,y1,y2

```python
# Selecting feature matrix(X) and prediction vector(y)

# selecting columns for X
X=dataset.iloc[:,2:12].values
# selecting columns for y Temprature
y=dataset.iloc[:,[12]].values 
# selecting columns for y Relative humidity
y1=dataset.iloc[:,[13]].values 
# selecting columns for y Absolute humidity
y2=dataset.iloc[:,[14]].values 
```

### Preprocessing the data

Missing values are tagged with -200 value

```python
# Replacing Misssing values
from sklearn.impute import SimpleImputer
sim = SimpleImputer(missing_values=-200, strategy='mean')
X = sim.fit_transform(X)
X
```

```python
# Removing rows where y is missing as their result would be misleading
missin_value_rows=np.where(y==-200) # Geeting rows with missing values
missin_value_rows[0]
```

```python
# deleting from y
y=np.delete(y,missin_value_rows[0]) 
# deleting from y1
y1=np.delete(y1,missin_value_rows[0]) 
# deleting from y2
y2=np.delete(y2,missin_value_rows[0]) 
# Deleting from X also axis is given as it's a matrix
X=np.delete(X,missin_value_rows[0], axis=0) 
```

```python
# Scaling the data using the standard Scaler

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

Applying Linear Regression model

```python
# 3 Objects of LinearRegression for 3 differen valuess(Temprature, Relative humidity, Absolute humidity)

from sklearn.linear_model import LinearRegression
# LR for y
lin_reg=LinearRegression() 
# LR for y1
lin_reg1=LinearRegression() 
# LR for y2
lin_reg2=LinearRegression() 
```

```python
# Fitting the model for Temprature
lin_reg.fit(X,y)
```

```python
# Fitting the model for Relative humidity
lin_reg1.fit(X,y1)
```

```python
# Fitting the model for Absolute humidity
lin_reg2.fit(X,y2)
```

```python
# Checking the prediction score for Temprature
lin_reg.score(X,y)
```

We get a score of 61% it is just fine

```python
# Checking the prediction score for Relative humidity
lin_reg1.score(X,y1)
```

We got a score of 44% which is bad

```python
# Checking the prediction score for Absolute humidity
lin_reg2.score(X,y2)
```

We got a score of 78% which is just fine.
